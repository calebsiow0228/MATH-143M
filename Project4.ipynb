```python
import numpy as np
from scipy.sparse import diags
```

```python
def matrix_A(n):
    k = [-1* np.ones(n-1), np.ones(n), -1 * np.ones(n-1)]
    offset = [-1, 0, 1]
    A = diags(k, offset).toarray()
    A = np.array(A)
    return A
```

```python
def x_guess(n):
    x = np.ones([n, 1])
    return x
```

```python
def power_method(Matrix, vector, TOL = 0.01):
    A = np.copy(Matrix)
    eigenvector = np.copy(vector)
    eigenvalue = 0
    iterations = 0
    conv = False
    while conv == False:
        x = np.dot(A, eigenvector)
        u = np.linalg.norm(x, np.inf)
        x = x/u
        error = np.linalg.norm((x - eigenvector), np.inf)
        if error <= TOL:
            conv = True
        iterations += 1
        eigenvector = x
        eigenvalue = u
    return eigenvalue, eigenvector, iterations
```

```python
def hotelling_deflation(A, eig_1, eig_vec_1):
    B = A - ((eig_1)/np.dot(eig_vec_1.T, eig_vec_1)) * np.dot(eig_vec_1, eig_vec_1.T)
    return power_method(B, eig_vec_1)
```

```python
if __name__ == '__main__':
    A = matrix_A(50)
    x0 = x_guess(50)
    eigenvalue_1, eigenvector_1, iterations = power_method(A, x0) #returns max eigenvalue, corresponding eigenvector to eigen max, and numbef of iterations using the power method
    eigenvalue_2, eigenvector_2, iterations_2 = hotelling_deflation(A, eigenvalue_1, eigenvector_1) #return second largest eigenvalue, corresponding eigenvector to max lam 2, and num of iter using deflation
    print("3. The Max Eigenvalue Is: ", np.around(eigenvalue_1, 4))
    print("4. The number of iterations required to converge for power method: ", iterations)
    print("5. Second Largest Eigenvalue using hotelling deflation: ", np.around(eigenvalue_2, 4))
    print("6. The number of iterations required to converge using hotelling deflation: ", iterations_2)
    
    
    
```
